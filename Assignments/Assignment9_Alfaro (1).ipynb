#### CASE Statements
Case statements are similar to the if-then-else statements in programming. The data that meets the case statements in the database will be returned. You start a CASE statement with the word CASE and conclude it with an END. Between those keywords, you specify each condition with a WHEN [condition] THEN [value], where the condition and the values are specified by you.

Nield, Thomas. Getting Started with SQL (p. 71). O'Reilly Media. Kindle Edition. <br>

**Example Question:** did customers purchase more than one item?
<br>
**What tables are we joining? (If any):** only using transaction table here

```python
# seeing if someone bought more than 1 item
# this is just checking if the quantity is bigger than one
query_case1 = """
SELECT 
    transaction_id,
    transaction_qty,
    CASE 
        WHEN transaction_qty > 1 THEN 'Yes'
        ELSE 'No'
    END AS more_than_one
FROM transactions
LIMIT 30;
"""
pd.read_sql(query_case1, engine)
```

**Question:** splitting most affordable to most expensive into two groups. whats affordable and whats expensive? what is the min & max of those groups?
<br>
**What table(s) are we joining?** 

```python
# combining CASE with min and max
# sorting products into affordable vs expensive
# then grabbing the min and max price inside each group

query_case_min_max = """
SELECT 
    CASE 
        WHEN unit_price > 10 THEN 'Expensive'
        ELSE 'Affordable'
    END AS price_category,
    MIN(unit_price) AS min,
    MAX(unit_price) AS max
FROM products
GROUP BY price_category;
"""

pd.read_sql(query_case_min_max, engine)
```

**MIN() and MAX()** <br>
AVG() will take the average of a numeric field.

**Question:**  What is the minimum, maximum, and average unit price of products?

<br>

```python
# finding the cheapest, most expensive, and average product prices
# MIN gets the lowest price, MAX gets the highest, AVG gets the average

query_min_max_avg = """
SELECT 
    MIN(unit_price) AS minimum_price,
    MAX(unit_price) AS maximum_price,
    AVG(unit_price) AS average_price
FROM products;
"""

pd.read_sql(query_min_max_avg, engine)

# average price is 7$
```

**Aliases (AS)**
<br>
You can abbreviate your code to make it more visually appealing...or more confusing? :) <br>
<br>
Examples:<br>
FROM table_name t<br>

<br>
**Objective:** Use abbreviations or aliases for all tables for the same code you wrote above (If you have not done so already). Be sure to obtain the same result set.

```python
#using p for products table

query_min_max_avg_alias = """
SELECT 
    MIN(p.unit_price) AS minimum_price,
    MAX(p.unit_price) AS maximum_price,
    AVG(p.unit_price) AS average_price
FROM products AS p;
"""

pd.read_sql(query_min_max_avg_alias, engine)
```
